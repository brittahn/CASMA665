### INF Model



dV/dt = I/C

Derive numerical solution

V(t+1) = V(t) + dt*(I/C)

```python
import numpy as np
import matplotlib.pyplot as plt

C=0.1                   # capcitance created by membrane which acts as an insulator
I=1.0                   # current
dt=0.01                 # (s) time step
Vth=-50                 # threshold voltage
Vreset=-70              # reset voltage
V = np.zeros([1000,1])  # initialize voltage array; 1000 rows, 1 col
V[0]=-70  # initial val of V

for i in range(0,999):
    V[i+1] = V[i] + dt*(I/C)
    if V[i+1]>Vth:      # reset voltage if above threshold
        V[i+1]=Vreset

t = np.arange(0, len(V))*dt  # arr of times

plt.figure()
plt.plot(t,V)
plt.title("Toy INF Model")
plt.xlabel('Time [s]')
plt.ylabel('V')
plt.show()
```
