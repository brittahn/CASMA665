```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define biological constants
C=1                      # capcitance
I=1                      # input current
dt=0.01                  # time step (s)
Vth=-70                  # threshold voltage (mV)
Vreset=-80               # reset voltage (mV)
gL=1                     # leak conductance
EL=-70                   # leak equilibrium
Vap=20                   # action potential (AP) max voltage

# Initialize voltage array
V = np.zeros([1000,1])   # voltages; 1000 rows, 1 col
V[0]=-70                 # initial voltage

# LIF Model
for i in range(0,999):
    V[i+1] = (-gL*(V[i]-EL) + I)*(dt/C) + V[i]  # numerical solution for LIF
    if V[i+1]>=Vth:      # if voltage equals or surpasses threshold,
        V[i]=Vap         # then spike,
        V[i+1]=Vreset    # then reset the voltage

# Create time array
t = np.arange(0, len(V))*dt

# Plot
plt.figure()
plt.plot(t,V)
plt.title("Challenge 1.1: LIF Model with Small Input Current")
plt.xlabel('Time (ms)')
plt.ylabel('V (mV)')
plt.axhline(y = Vth, color = 'r', linestyle = '-', label='Threshold Voltage')
legend = plt.legend(loc='upper right')
plt.ylim(Vreset + (-20), Vap + 20)
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

"""
Function Name: fi

Description: Frequency-Current (F-I) Curve.
Computes firing rate for each input current.

Args:
I -- input current

Returns:
firing_rate -- firing rate of LIF neuron at given input current level
"""
def fi(I):

    # Define biological constants
    C=.25                    # make capacitance to accentuate firing rates
    I=I                      # input current
    dt=0.01                  # time step (s)
    Vth=-70                  # threshold voltage (mV)
    Vreset=-80               # reset voltage (mV)
    gL=1                     # leak conductance
    EL=-70                   # leak equilibrium
    Vap=20                   # action potential (AP) max voltage
    
    # Initialize voltage array
    V = np.zeros([1000,1])   # voltages; 1000 rows, 1 col
    V[0]=-70                 # initial voltage

    # Create time array
    t = np.arange(0, len(V))*dt



    # Initialize spike count
    spike_count=0

    # LIF Model
    for i in range(0,999):
        V[i+1] = (-gL*(V[i]-EL) + I)*(dt/C) + V[i]  # numerical solution for LIF
        if V[i+1]>=Vth:      # if voltage equals or surpasses threshold,
            V[i]=Vap         # then spike,
            V[i+1]=Vreset    # then reset the voltage
            spike_count+=1   # add to spike count
    firing_rate=spike_count/t[-1]  # compute firing rate

    return firing_rate

# Create array of currents
currents=[]
curr_init=0
for x in range(100):
    curr_init+=.1
    currents.append(curr_init)

# Call function to create array of frequencies
freq=[]
for i in currents:
    freq.append(fi(i))

# Plot FI curve
plt.figure()
plt.plot(currents, freq, color='black')
plt.title("F-I Curve of LIF Neuron")
plt.xlabel("Input Current")
plt.ylabel("Firing Rate")
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Define functions
"""
Descr: dvdt is a function that computes the change in voltage over time, explained by the Izhikevich neuron model.

Eq: dvdt = 0.04*(v**2) + 5*v + 140 - u + I

Args: v, u, I
u --- membrane recovery variable
v --- membrane potential
I --- injected current

Returns: 0.04*(v**2) + 5*v + 140 - u + I
"""
def dvdt(v, u, I):
    return (0.04*(v**2) + 5*v + 140 - u + I)


"""
Descr: dudt is a function that computes the change in membrane recovery over time, explained by the Izhikevich neuron model.

Eq: dudt = a*(b*v - u)

Args: a, b, v, u
a --- time scale of recovery variable u. smaller values of a => slower recovery. typically a=0.02
b --- sensitivity of recovery variable u to subthreshold fluctuations of membrane potential v
v --- membrane potential
u --- membrane recovery variable

Returns:
"""
def dudt(a, b, v, u):
    return (a*(b*v - u))


# Define variables
I=4         # injected current
a=0.02      # time scale of recovery variable u
b=0.2
c=-65       # mV
d=8
t_sim=2000  # simulation time
dt=0.01     # time step


# Initialize t, u, v arrays
t_values=np.arange(0, t_sim, dt)  # times
u_values=np.zeros(t_values.size)  # empty array of membrane recovery values
v_values=np.zeros(t_values.size)  # empty array of voltages
v_values[0]=-75                   # initial voltage (mV)
u_values[0]=0                     # initial membrane recovery value


# Build arrays of voltages and membrane recovery values.
for i in range(t_values.size - 1):

    dv = dvdt(v_values[i], u_values[i], I)
    du = dudt(a, b, v_values[i], u_values[i])
    v_values[i+1] = (v_values[i] + dv*dt)
    u_values[i+1] = (u_values[i] + du*dt)
    
    if v_values[i+1]>=30:  # if v>=30 mV, => reset s.t. (v = c) and (u = u+d)
        v_values[i+1] = c
        u_values[i+1] = u_values[i]+d

# Plot voltage trace
plt.figure(figsize=(10,6))
plt.plot(t_values, v_values)
plt.title("Izhikevich Model - Excitatory cortical cells, regular spiking: Membrane Potential")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.show()

plt.figure(figsize=(10,6))
plt.plot(t_values, u_values)
plt.title("Izhikevich Model - Excitatory cortical cells, regular spiking: Membrane Recovery Variable")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Recovery Variable u")
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define constants
C=1.0                   # capcitance
I=1.0                   # current
R=1.0
dt=0.01                 # time step (s)
Vth=-50               # threshold voltage
Vreset=-75              # reset voltage
V = np.zeros([1000,1])  # initialize voltage array; 1000 rows, 1 col
V[0]=-70                # initial val of V
Vap=20

# Create voltage array with LIF model equation
for i in range(0,999):
    V[i+1] = (-(V[i]-(I*R))/(R*C))*dt + V[i]
    if V[i+1]>=Vth:      # if voltage surpasses threshold, then spike and reset voltage.
        V[i]=Vap
        V[i+1]=Vreset    # reset voltage

# Create an array of times
t = np.arange(0, len(V))*dt

# Plot
plt.figure()
plt.plot(t,V)
plt.title("Challenge 3: LIF Model with Resistance=1")
plt.xlabel('Time (s)')
plt.ylabel('V (mV)')
plt.axhline(y = Vth, color = 'r', linestyle = '-', label='Threshold Voltage') # threshold voltage
legend = plt.legend(loc='upper right')
plt.ylim(Vreset + (-20), Vap + 20)
plt.show()
```
