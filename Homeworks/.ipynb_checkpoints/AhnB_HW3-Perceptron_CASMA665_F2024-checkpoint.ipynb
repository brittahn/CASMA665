## <span style="color:cornflowerblue"><ins>**Challenge 1**</ins></span>


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the file training_data.csv
training_data = pd.read_csv("~/Desktop/CASMA665/Homeworks/training_data.csv")
print(training_data)

# Save data to variables
x_train = training_data.x_training
y_train = training_data.y_training
correct_train = training_data.correct_answer
```

```python
# Plot the solutions of the data points classified as above and below
x_train_above=[]
y_train_above=[]
x_train_below=[]
y_train_below=[]

for i in range(x_train.size):
    if correct_train[i] > 0: # if 1, it's above the line
        x_train_above.append(x_train[i])
        y_train_above.append(y_train[i])
    else:
        x_train_below.append(x_train[i])
        y_train_below.append(y_train[i])

plt.figure()
plt.scatter(x_train_above, y_train_above)
plt.scatter(x_train_below, y_train_below)
plt.show()

print("Number of testing data points above the line: ", len(x_train_above))
print("Number of testing data points below the line: ", len(x_train_below))
```

```python
# Find the equation of the line which separates the above and below data points.

# Define initial values
slope = 1
intercept = 1
wx = 1
wy = 1
wb = 1
bias = 1
learning_constant = .01
estimated_slope = np.zeros(x_train.size)
estimated_intercept = np.zeros(x_train.size)

"""
Function Name: feedforward

Description: Function that computes the perceptron's solution

Args:
x  --- x coordinate of given point
y  --- y coordinate of given point
wx --- weight of x
wy --- weight of y
wb --- weight of bias term

Returns:
1  --- perceptron's guess is that the point is above the line
0  --- perceptron's guess is that the point is below the line
"""
def feedforward(x,y,wx,wy,wb):
    if (x*wx + y*wy + wb)>0:
        return 1
    else:
        return 0

# Train perceptron to find equation for the line
for i in range(x_train.size):  # from 0 to 1999

    # For each data point in given training data set (that has correct solutions),
    x = x_train[i]
    y = y_train[i]

    # Error = known_solution - perceptron's guess
    error = correct_train[i] - feedforward(x, y, wx, wy, wb)

    # If there is an error, update the weights based on the learning constant
    if error != 0:
        wx = wx + error * x * learning_constant
        wy = wy + error * y * learning_constant
        wb = wb + error * bias * learning_constant

    # Compute the estimated slope from perceptron.
    estimated_slope[i] = -wx/wy;

    # Compute the estimated intercept from perceptron.
    estimated_intercept[i] = -wb/wy;


####
# Display results

# Create a range for x values in plot
x_range = np.linspace(-5,5,100);

# Plot perceptron guesses with all data points
for k in range(1,2000,100):
    plt.plot(x_range, estimated_slope[k]*x_range+estimated_intercept[k], 'purple', alpha=0.25)

plt.title('Perceptron guesses with training data ')
plt.scatter(x_train_above, y_train_above)
plt.scatter(x_train_below, y_train_below)
plt.show()

# Plot final line
final_slope = estimated_slope[-1]
final_intercept = estimated_intercept[-1]
plt.figure()
plt.title('Final perceptron guess with training data ')
plt.plot(x_range, final_slope*x_range + final_intercept, 'black')
plt.scatter(x_train_above, y_train_above)
plt.scatter(x_train_below, y_train_below)
plt.show()

print("The equation for the line is (y=mx+b): y = ", final_slope, "x + ", final_intercept)
```

#### Use the perceptron trained on the training data to analyse the testing data.

```python
# Load the file testing_data.csv
testing_data = pd.read_csv("~/Desktop/CASMA665/Homeworks/testing_data.csv")
print(testing_data)

x_test = testing_data.x_testing
y_test = testing_data.y_testing
```

```python
# Create empty lists to store points above or below line
x_test_above = []
y_test_above = []
x_test_below = []
y_test_below = []
correct_test = []

for i in range(x_test.size): # from 0 to 99
    if (y_test[i]) > (final_slope*x_test[i] + final_intercept):  # if y of test point is greater than y of perceptron line,
        x_test_above.append(x_test[i])                           # ... then this point is above the line
        y_test_above.append(y_test[i])                           # ... so add it to the list of x and y points above line.
        correct_test.append(1)                                   # correct answer is 1
    else:
        x_test_below.append(x_test[i])                           # ... then this point is above the line
        y_test_below.append(y_test[i])                           # ... so add it to the list of x and y points below line.
        correct_test.append(0)                                   # correct answer is 0

# Recall: perceptron's final guess for the equation of the line is:
# y =  0.5914784345578687 x +  -1.363865088029324
plt.figure()
plt.plot(x_range, final_slope*x_range + final_intercept, 'black', label='Perceptron Line')
plt.scatter(x_test_above, y_test_above)
plt.scatter(x_test_below, y_test_below)
plt.legend()
plt.show()

print("Number of testing data points above the line: ", len(x_test_above))
print("Number of testing data points below the line: ", len(x_test_below))
```
